```python
import numpy as np
import matplotlib.pyplot as plt
import random
pi = 3.14159265359
```

```python
def plott(dmax, step):
  d = np.zeros(0) #distances accordind to the population density of the galaxies
  ep = 0.005
  for i in range(5, dmax + 1, step):
    m = ((i+ep)**3-(i-ep)**3)/((5+ep)**3-(5-ep)**3) #ratio of number of galaxies in specific region to that in initial region(5)
                                                    #where i have taken one galaxy
    m = int(m)
    for _ in range(m):                                       #assigning that many galxies in that region
      d = np.append(d, random.uniform(i-step/2, i + step/2)) #the distances of the galaxies , i have distributed acoording to the
                                                             #number density and spread the distances randomly around the specific 
                                                             #distance to make the graph more continous 
  n = d.size
  sigma = 500 #standard deviation
  vr = np.random.normal(0, sigma, n) #random velocity
  v = np.zeros(n)
  v0 = 70*d #velocity due to hubbles law
  theta = 2*pi*np.random.rand(n) #angle between random veloxity and v0 
  v = np.sqrt(v0**2 + vr**2 - 2*v0*np.cos(theta)*vr ) #actual velocity
  plt.plot(d, v, 'o')
  plt.plot(d, v0)
  plt.show()
```

```python
plott(30, 2)
```

```python
plott(50, 5)
```

```python
plott(75, 10)
```
