```python
import numpy as np
import matplotlib.pyplot as plt
pi = 3.14159265359
```

```python
r1, r2 = 5, 2.5 #radius
n = 300 #to represent total number of points to calculate intesity and also to denote number of time values i am taking to draw the graph
points = np.zeros([n,n,3]) #refers to the array of points ,their coordinates and whether they are on or off
for x in range(n):   # assigning the on/off
  for y in range(n):
    points[x][y][0] = (x-n/2) * (r1/n)*2
    points[x][y][1] = (y-n/2) * (r1/n)*2
    points[x][y][2] = int(np.sqrt(points[x][y][0]**2+points[x][y][1]**2)<=r1)
t = np.array([(x-n/2) * ((r1+r2+2)/n)*2 for x in range(n)])  #these denote the time value of transit
intensity = np.zeros(n)
```

```python
def get_i(p): #to get the intensity from the array of points
  sum = 0
  for x in range(n):
    for y in range(n):
      sum = sum + p[x][y][2]
  return sum

def update_i(p, t): #update the on/off values of the grid points
  for x in range(n):   
    for y in range(n):
      points[x][y][2] = int(np.sqrt(points[x][y][0]**2+points[x][y][1]**2)<=r1 and np.sqrt((points[x][y][0]-t)**2+points[x][y][1]**2)>r2)
```

```python
for i in range(n):
  update_i(points, t[i])
  intensity[i] = get_i(points)

plt.plot((t+10)*10, intensity)
plt.show()
```
